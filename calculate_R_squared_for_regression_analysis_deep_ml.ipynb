R-squared also known as the coefficient of determenation, is a measure that indicates how well the independent variables explain the variability (variance) of the dependent variable in regression model
implement a python function that calculates the R-squared value, given arrays of tyrue values y_true and predicted values y_pred

R² = 1 means the model perfectly explains the variance in the dependent variable.
R² = 0 means the model explains none of the variance.

image.png

```python
def r_squared(y_true,y_pred):
  if np.array_equal(y_true,y_pred):
    return 1.0
  res=np.sum((y_true-y_pred)**2)
  tot=np.sum((y_true-np.mean(y_true))**2)
  try:
    r_squared=1-(res/tot)
    if np.isinf(r_squared):
      return 0.0
    return np.round(r_squared,3)
  except ZeroDivisionError:
    return 0.0
```

```python
import numpy as np
y_true=np.array([1,2,3,4,5])
y_pred=np.array([1.1,2.1,2.9,4.2,4.8])
print(r_squared(y_true,y_pred))
```
